```python
def select_duplicate_examples(df, sub_label, obj_label):
    """
    Seleciona os dois primeiros exemplos de cada combinação duplicada de 'sub_label' e 'obj_label' em um DataFrame.

    Parâmetros:
    df (pandas.DataFrame): O DataFrame contendo as colunas 'sub_label', 'obj_label' e 'masked_sentence'.

    Retorna:
    pandas.DataFrame: Um DataFrame contendo apenas os dois primeiros exemplos de cada combinação duplicada.
    """
    # Contar o número de ocorrências de cada combinação de sub_label e obj_label
    counts = df.groupby([sub_label, obj_label]).size()

    # Selecionar apenas as combinações que possuem mais de 1 ocorrência
    duplicate_combinations = counts[counts > 1].index.tolist()

    # Criar um DataFrame contendo apenas essas combinações duplicadas
    duplicated_df = df[df.apply(lambda row: (row[sub_label], row[obj_label]) in duplicate_combinations, axis=1)]

    # Selecionar apenas os dois primeiros exemplos de cada combinação duplicada
    final_df = duplicated_df.groupby([sub_label, obj_label]).head(2).reset_index(drop=True)

    return final_df
```
